```python
data["area_per_bed"] = data["area"] / data["bedrooms"]
```

```python
data2 = data[data["area_per_bed"] >=500]
```

```python
data2["price_per_area"] = round(data2["price"]*100000 / data["area"],2)
```

```python
data3 = data2[data2["price_per_area"] >2.703956e+07]
```

```python
data3.drop(columns=["area_per_bed","price_per_area"],axis=1,inplace=True)
```

```python
data3
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer

```

```python
col_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ["mainroad", "guestroom"]),
    remainder="passthrough"
)
```

```python
lr = LinearRegression()
```

```python
scaler = StandardScaler()
```

```python
model = make_pipeline(col_trans,lr)
```

```python
input_data = data3.drop(columns="price")
output_data = data3["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
input = pd.DataFrame([[7240,4,2,"yes","no"]],columns=["area","bedrooms","bathrooms","mainroad","guestroom"])
```

```python
prediction = model.predict(input)
print(round(prediction[0], 0))
```
